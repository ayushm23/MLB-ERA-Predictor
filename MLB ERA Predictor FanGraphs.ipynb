```python
import numpy as np
import pandas as pd
import re
import requests
from bs4 import BeautifulSoup

from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
from tensorflow.keras.optimizers import SGD

#https://machinelearningmastery.com/regression-tutorial-keras-deep-learning-library-python/
#https://machinelearningmastery.com/how-to-make-classification-and-regression-predictions-for-deep-learning-models-in-keras/
```

```python
def get_data():
    df = pd.read_csv('FanGraphs Pitching Leaderboard.csv', header=0, index_col = False)
    df = df.drop(['Team', 'W', 'L', 'SV', 'G', 'GS', 'LOB%', 'GB%', 'HR/FB', 'FIP', 'xFIP', 'playerid'], axis=1)
    return df

def split_data(df):
    is_2018 = df['Season'] == 2018
    is_2019 = df['Season'] == 2019
    before_2018 = df['Season'] < 2018
    train = df[before_2018]
    test = df[is_2018]
    compare = df[is_2019]
    return train, test, compare

df = get_data()
#df.head()

train_df, test_df, compare_df = split_data(df)
```

```python
def train_model(df):
#     print(df)
    X = df.drop(['ERA', 'Name', 'Season'], axis=1)
    Y = df[['ERA']]
#     print(X)
#     print(Y)

    model = Sequential()
    numCols = len(X.columns)
    #print(numCols)
    model.add(Dense(8, input_dim=numCols, kernel_initializer = 'ones', activation='relu'))
    model.add(Dense(4, kernel_initializer = 'ones', activation='relu'))
    model.add(Dense(1, kernel_initializer = 'ones', activation='linear'))
        
    model.compile(loss='mse', optimizer='adam', metrics=['mse'])
    model.fit(X, Y, epochs=1000)
    return model
     
era_model = train_model(train_df)
```

```python
def predict_model(df, model):
    #     print(df)
    X = df.drop(['ERA', 'Name', 'Season'], axis=1)
    Y = df[['ERA']]
    Y = Y.values
#     print(X)

    preds = model.predict(X)
    return (Y, preds)

def test_model(df, model):
    #     print(df)
    X = df.drop(['ERA', 'Name', 'Season'], axis=1)
    Y = df[['ERA']]
#     print(X)
#     print(Y)

    #outputs the MSE
    return(model.evaluate(X, Y, batch_size=64))
```

```python
# test_model(test_df, era_model)
actual, preds = predict_model(test_df, era_model)

actual = list(map(float, actual))
preds = preds.flatten().tolist()
preds = [round(n, 2) for n in preds]

final_df = test_df
final_df['Predicted ERA'] = preds
final_df['2019 ERA'] = np.nan
final_df.set_index('Name', inplace=True)
#print(final_df)
# compare_df.set_index('Name')

for name, row in final_df.iterrows():
    if name in compare_df.values:
        era_2019 = compare_df.loc[compare_df['Name'] == name]['ERA'].values[0] #type is 'pandas.core.series.Series'
        final_df.at[name, '2019 ERA'] = era_2019

final_df.to_csv('ERA Predictions 2018.csv', index=True)
#compare_df.to_csv('Fangraphs Pitching 2019.csv', index=False)
```
